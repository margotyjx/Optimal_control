```python
import numpy as np
import torch
from torch.autograd import Variable
import torch.nn as nn
import pandas as pd
import matplotlib.pyplot as plt
import numpy 
from pathlib import Path
import math
```

```python
# Original potential function
class functions:
    def __init__(self,beta,alpha):
        self.beta = beta
        self.alpha = alpha
        
    def funU(self,x):
        U = 0.5*self.beta*(x**2) + 1/4*self.alpha*(x**4)

        return U

    def funT(self,p):
        T = 0.5*(p**2)
        return T

    def dU(self,x):
        dUx = x*(self.beta + self.alpha * (x**2))

        return dUx
```

```python
class Model1(nn.Module):
    """Feedfoward neural network with 1 hidden layer"""
    def __init__(self, in_size, hidden_size, out_size):
        super().__init__()
        # hidden layer
        self.linear1 = nn.Linear(in_size, hidden_size)
        # output layer
        self.linear2 = nn.Linear(hidden_size, out_size)
        
    def forward(self, xb):
        # Get intermediate outputs using hidden layer
        out = self.linear1(xb)
        # Apply activation function
        tanhf = nn.Tanh()
        out = tanhf(out)
        # Get predictions using output layer
        out = self.linear2(out)
        # apply activation function again
        out = torch.sigmoid(out)
        return out
    
class Model2(nn.Module):
    """Feedfoward neural network with 1 hidden layer"""
    def __init__(self, in_size, hidden_size, out_size):
        super().__init__()
        # hidden layer
        self.linear1 = nn.Linear(in_size, hidden_size)
        # output layer
        self.linear2 = nn.Linear(hidden_size, hidden_size)
        self.linear3 = nn.Linear(hidden_size, out_size)
        
    def forward(self, xb):
        # Get intermediate outputs using hidden layer
        out = self.linear1(xb)
        # Apply activation function
        tanhf = nn.Tanh()
        out = tanhf(out)
        # Get predictions using output layer
        out = self.linear2(out)
        out = tanhf(out)
        out = self.linear3(out)
        # apply activation function again
        out = torch.sigmoid(out)
        return out
class Model3(nn.Module):
    """Feedfoward neural network with 1 hidden layer"""
    def __init__(self, in_size, hidden_size, out_size):
        super().__init__()
        # hidden layer
        self.linear1 = nn.Linear(in_size, hidden_size)
        # output layer
        self.linear2 = nn.Linear(hidden_size, hidden_size)
        self.linear3 = nn.Linear(hidden_size,hidden_size)
        self.linear4 = nn.Linear(hidden_size, out_size)
        
    def forward(self, xb):
        # Get intermediate outputs using hidden layer
        out = self.linear1(xb)
        # Apply activation function
        tanhf = nn.Tanh()
        out = tanhf(out)
        # Get predictions using output layer
        out = self.linear2(out)
        out = tanhf(out)
        out = self.linear3(out)
        out = tanhf(out)
        out = self.linear4(out)
        # apply activation function again
        out = torch.sigmoid(out)
        return out
```

# compare with FEM, beta = 10

```python
data_folder = Path('./Duffing_beta10')
pts_file = data_folder/'Duffing_pts_ellip.csv'
q_file = data_folder/'Duffing_committor_beta10_ellip.csv'
qminus_file = data_folder/'Duffing_backward_committor_beta10_ellip.csv'

data_pts = np.loadtxt(pts_file, delimiter=',', dtype=float)
data_q = np.loadtxt(q_file, delimiter=',', dtype=float)
data_qminus = np.loadtxt(qminus_file,delimiter=',', dtype=float)
data_pts_minus = np.hstack((data_pts[:,0][:,None],-data_pts[:,1][:,None]))

pts_beta10 = torch.tensor(data_pts,dtype = torch.float32)
pts_beta10_minus = torch.tensor(data_pts_minus,dtype = torch.float32)
q_fem_beta10 = torch.tensor(data_q[:,2],dtype = torch.float32)
qminus_fem_beta10 = torch.tensor(data_qminus[:,2],dtype = torch.float32)
```

```python
model_beta10 = torch.load('Duffiing_gamma0.5_beta10_PINN_EM.pt')
model_beta10.eval()
q_NN_beta10 = model_beta10(pts_beta10)

plt.scatter(data_pts[:,0], data_pts[:,1],c = q_NN_beta10.detach().numpy(), s = 2)
# plt.title('estimated committor by NN')
plt.colorbar()
plt.xlabel('$x$')
plt.ylabel('$p$')
plt.title('$q^+$')
plt.savefig(data_folder/'forward_NN_beta10_ellip.pdf')
plt.show()

plt.scatter(data_pts[:,0], data_pts[:,1],c = q_fem_beta10.detach().numpy(), s = 2)

plt.colorbar()
plt.xlabel('$x$')
plt.ylabel('$p$')
plt.title('$q^+$')
plt.savefig(data_folder/'forward_FEM_beta10_ellip.pdf')
plt.show()

plt.scatter(data_pts[:,0], data_pts[:,1],
           c = torch.abs(q_fem_beta10[:,None] - q_NN_beta10).detach().numpy(), s = 2)
plt.title('absolute difference between FEM and NN of forward committor')
plt.colorbar()
plt.show()

MAE = torch.mean(torch.abs(q_NN_beta10 - q_fem_beta10[:,None]))
RMSE = torch.sqrt(torch.mean((q_NN_beta10 - q_fem_beta10[:,None])**2))
print('absolute MAE forward is: ',MAE)
print('absolute RMSE forward is : ',RMSE)
```

```python
# Compare the result of backward committor function
qminus_NN_beta10 = torch.tensor(1) - model_beta10(pts_beta10_minus)

plt.scatter(data_pts[:,0], data_pts[:,1],c = qminus_NN_beta10.detach().numpy(), s = 2)
# plt.title('estimated committor by NN')
plt.colorbar()
plt.xlabel('$x$')
plt.ylabel('$p$')
plt.title('$q^-$')
plt.savefig(data_folder/'backward_NN_beta10_ellip.pdf')
plt.show()

plt.scatter(data_pts[:,0], data_pts[:,1],c = qminus_fem_beta10.detach().numpy(), s = 2)
# plt.title('estimated committor by FEM')
plt.colorbar()
plt.xlabel('$x$')
plt.ylabel('$p$')
plt.title('$q^-$')
plt.savefig(data_folder/'backward_FEM_beta10_ellip.pdf')
plt.show()

plt.scatter(data_pts[:,0], data_pts[:,1],
           c = torch.abs(qminus_fem_beta10[:,None] - qminus_NN_beta10).detach().numpy(), s = 2)
plt.title('absolute difference between FEM and NN of forward committor')
plt.colorbar()
plt.show()

MAE = torch.mean(torch.abs(qminus_NN_beta10 - qminus_fem_beta10[:,None]))
RMSE = torch.sqrt(torch.mean((qminus_NN_beta10 - qminus_fem_beta10[:,None])**2))
print('absolute MAE backward is: ',MAE)
print('absolute RMSE backward is : ',RMSE)
```

```python
Funs = functions(torch.tensor(-1), torch.tensor(1))
gamma = torch.tensor(0.5)
betaT = torch.tensor(10)

U = Funs.funU(pts_beta10[:,0]) # potential energy function
T = Funs.funT(pts_beta10[:,1]) # kinetic energy function

normal_Z = torch.sum(torch.exp(-betaT*(U+T))*q_fem_beta10*qminus_fem_beta10)
mR_beta10 = torch.exp(-betaT*(U+T))*q_fem_beta10*qminus_fem_beta10/normal_Z
```

```python
diff_beta10 = torch.abs(q_fem_beta10 - q_NN_beta10.squeeze())
Weighted_MAE_beta10 = torch.sum(mR_beta10*diff_beta10)
Weighted_RMSE_beta10 = torch.sqrt(torch.sum(mR_beta10*diff_beta10.pow(2)))

print('weighted MAE forward: ', Weighted_MAE_beta10)
print('weighted RMSE forward: ', Weighted_RMSE_beta10)

diff_minus_beta10 = torch.abs(qminus_fem_beta10 - qminus_NN_beta10.squeeze())
Weighted_MAE_beta10 = torch.sum(mR_beta10*diff_minus_beta10)
Weighted_RMSE_beta10 = torch.sqrt(torch.sum(mR_beta10*diff_minus_beta10.pow(2)))

print('weighted MAE backward: ', Weighted_MAE_beta10)
print('weighted RMSE backward: ', Weighted_RMSE_beta10)
```

```python
plt.scatter(data_pts[:,0], data_pts[:,1],
           c = torch.abs(mR_beta10*diff_beta10).detach().numpy(), s = 2)
plt.title('weighted difference between FEM and NN results for forward committor')
plt.colorbar()
plt.show()
```

```python
plt.scatter(data_pts[:,0], data_pts[:,1],
           c = torch.abs(mR_beta10*diff_minus_beta10).detach().numpy(), s = 2)
plt.title('weighted difference between FEM and NN results for backward committor')
plt.colorbar()
plt.show()
```
